```python
import os
import rasterio

input_folder = "/Volumes/External/TJ/020_preprocessing"
output_folder = "/Volumes/External/TJ/020_preprocessing/sentinel1_1"

os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):

    # Skip macOS AppleDouble metadata files BEFORE trying to open
    if filename.startswith("._"):
        continue

    # Only process GeoTIFFs
    if not filename.lower().endswith((".tif", ".tiff")):
        continue

    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(
        output_folder,
        filename.replace(".tif", ".tif")
    )

    with rasterio.open(input_path) as src:
        if src.count < 2:
            print(f"Skipping {filename}: only has {src.count} band")
            continue

        profile = src.profile.copy()
        data = src.read()

        # Swap band 1 and band 2
        data[[0, 1]] = data[[1, 0]]

        with rasterio.open(output_path, "w", **profile) as dst:
            dst.write(data)

    print(f"Saved: {output_path}")
```
